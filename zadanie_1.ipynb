```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

Z populacji generalnej o rozkładzie normalnym $N(µ, 0.2)$ pobrano próbę (plik lista8_zad1.txt). Na poziomie istotności $α = 0.05$ zweryfikuj hipotezy
- $µ \ne 1.5$,
- $µ > 1.5$,
- $µ < 1.5$.

Narysuj odpowiednie obszary kytyczne i wyznacz p-wartości dla każdej z powyższych hipotez.
Odpowiedz na pytanie co stanie się kiedy zwiększymy bądź zmniejszymy poziom ufności.

```python
file = open('lista8_zad1.txt', 'r')
data = file.read().split()
file.close()

data = [float(x) for x in data]

mean = np.mean(data)
s = np.std(data)
n = len(data)

print("średnia: ", mean)
print("odchylenie standardowe: ", s)
print("n: ", n)
```

```python
mu0 = 1.5

T = (mean - mu0) / (np.sqrt(2) / np.sqrt(1000))
print(T)
```

```python
z_value = -3.18
p_value_two_tail = 2 * (1 - norm.cdf(abs(z_value)))
p_value_one_tail_right = norm.cdf(z_value)
p_value_one_tail_left = norm.cdf(z_value)

print("P-wartość dla dwustronnego testu: ", p_value_two_tail)
print("P-wartość dla jednostronnego testu (H1: mu > 1.5): ", p_value_one_tail_right)
print("P-wartość dla jednostronnego testu (H1: mu < 1.5): ", p_value_one_tail_left)
```

```python
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)
alpha = 0.05

z1 = norm.ppf(1 - alpha/2)
z2 = norm.ppf(1 - alpha)
```

```python
plt.plot(x, y, label='N(0,1)')
plt.fill_between(x, 0, y, where=(x < -z1) | (x > z1), color='red', alpha=0.5)
plt.axvline(-z1, color='red', linestyle='--', label='$Z_{1-\\frac{\\alpha}{2}} = -1.96$')
plt.axvline(z1, color='red', linestyle='--', label='$Z_{1+\\frac{\\alpha}{2}} = 1.96$')
plt.axvline(-3.18, color='g', linestyle='--', label='Z = -3.18')
plt.title(f'$H_1: \\mu \\ne 1.5$')
plt.legend(fontsize=8)
plt.savefig("krytyczny_11")
plt.show()
```

```python
plt.plot(x, y, label='N(0,1)')
plt.fill_between(x, 0, y, where=(x < -z2), color='red', alpha=0.5)
plt.axvline(-z2, color='red', linestyle='--', label='$-Z_{1-\\alpha} = -1.645$')
plt.axvline(-3.18, color='g', linestyle='--', label='Z = -3.18')
plt.title(f'$H_1: \\mu < 1.5$')
plt.legend()
plt.savefig("krytyczny_12.png")
plt.show()
```

```python
plt.plot(x, y, label='N(0,1)')
plt.fill_between(x, 0, y, where=(x > z2), color='red', alpha=0.5)
plt.axvline(z2, color='red', linestyle='--', label='$Z_{1-\\alpha} = 1.645$')
plt.axvline(-3.18, color='g', linestyle='--', label='Z = -3.18')
plt.title(f'$H_1: \\mu > 1.5$')
plt.legend()
plt.savefig("krytyczny_13.png")
plt.show()
```
